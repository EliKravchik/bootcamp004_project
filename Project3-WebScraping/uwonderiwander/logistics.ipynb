```python
from bs4 import BeautifulSoup
from decimal import *
import requests
import requests
import numpy as np
import pandas as pd
import unicodedata as ud
import re
import urllib2
import codecs
import math
```

```python
def save_string_field(dict, key, value, def_str, preamble):
    if len(value) > 0:
        if preamble == "":
            dict[key] = ud.normalize('NFKD', value).encode('ascii', 'ignore')
        else:
            dict[key] = preamble+value
    else: 
        dict[key] = def_str

def save_numeric_field(dict, key, value):
    if len(value) > 0:
        dict[key] = int(ud.normalize('NFKD', value).encode('ascii', 'ignore').replace(',', ''))
    else: 
        dict[key] = 0

def save_percent_field(dict, key, value):
    if len(value) > 0:
        dict[key] = int(ud.normalize('NFKD', value).encode('ascii', 'ignore').replace('%', ''))
    else: 
        dict[key] = 0

def save_millions_field(dict, key, value):
    nonBreakSpace = u'\xa0'
    value = value.strip(nonBreakSpace)
    value = value.strip('e')
    if len(value) > 0:
        dict[key] = int(ud.normalize('NFKD', value).encode('ascii', 'ignore').replace(',', '').\
            replace('$','').replace(' million',''))
    else: 
        dict[key] = 0

response = urllib2.urlopen('http://www.ttnews.com/top50/logistics/')
html = response.read()
soup = BeautifulSoup(html)
logistics_columns = ['Rank2015', 'Rank2014', 'CompanyName', 'Link', \
                    'NetRevenue', 'GrossRevenue', 'Employees']
logistics = pd.DataFrame(data=np.zeros((0,len(logistics_columns))), columns=logistics_columns)
logistics_table = soup.find(id= "rankings_box")
for row in logistics_table.findAll('tr'):
    col = row.findAll('td')
    data_dict = {}
    j = 0
    for field in col:
        data_dict["Link"] =  ""
        if j == 0:
            if field.get_text().isnumeric():
                save_numeric_field(data_dict, "Rank2015", field.get_text())
            else:
                break
        elif j == 1:
            save_numeric_field(data_dict, "Rank2014", field.get_text())
        elif j == 2:
            save_string_field(data_dict, "CompanyName", field.get_text(), "Unknown","")
        elif j == 3:
            save_millions_field(data_dict, "NetRevenue", field.get_text())
        elif j == 4:
            save_millions_field(data_dict, "GrossRevenue", field.get_text())
        else :
            save_numeric_field(data_dict, "Employees", field.get_text())
        j = j + 1
    single = pd.DataFrame([data_dict], columns=logistics_columns)
    print single
    logistics = logistics.append(single)
logistics.to_csv('logistics.csv')
logistics.value=nest_lst
```
